###Boolean selection of glucose lower than 180 and/& greater than 80


```python
df['inrange'] = (df['glucose'] < 180) & (df['glucose'] > 80)
```

```python
#rolling_sum
window = 30.5*288
inrange = pd.rolling_sum(df.inrange,window)
inrange = inrange.dropna()
inrange = inrange/float(window)

figure(figsize=(9,8))
#plot
inrange.plot()
```

###Visualize highs and lows of the day with masked arrays

```python
day = df.ix['2010-10-04']
highs = day[day['glucose'] > 180]
lows = day[day['glucose'] < 80]

figure(figsize=(9,6))
ax = gca()

day['glucose'].plot(style='k--',ax=ax)
highs['glucose'].plot(style='ro',ax=ax)
lows['glucose'].plot(style='bo',ax=ax)
```

#Indexing by built-in logic time chunks (micro, milli, sec, minute, hour, day, week, month)

Investigate averages per hour, per day, etc

```python
hour = df.index.hour
hour_g = df.groupby(hour).mean()
hour_g.plot(kind='bar')

day = df.index.day
day_g = df.groupby(day).mean()
day_g.plot(kind='bar')
```

#Resampling 
Investigate non-standard time-chunks: plot every 35 minutes

* Day Of Week Indexing
* Weekend warriors 

```python
days = days = pd.DateOffset(days=1)
for i in range(1,5):
    df[dt.datetime(2010,10,04,0,0,0)+days*(i-1):dt.datetime(2010,10,04,0,0,0)+days*i].plot(subplots=True)
```

```python
df_1min = df.resample('1Min',fill_method='pad')
min35 = pd.DateOffset(minutes=35)
for i in range(1,10):
    df_1min[dt.datetime(2010,10,10,0,0,0)+min35*(i-1):dt.datetime(2010,10,10,0,0,0)+min35*i].plot(subplots=True)
```

Computational Tools:

* rolling_count	Number of non-null observations
* rolling_sum	Sum of values
* rolling_mean	Mean of values
* rolling_median	Arithmetic median of values
* rolling_window	Moving window function
* ...

```python
#rolling averages
fig, ax = plt.subplots()
df.ix['2010-10-04'].plot(style='k--',ax=ax)
pd.rolling_mean(df.ix['2010-10-04'],10).plot(style='g',ax=ax)
```

```python
bolus_df = pd.read_csv('CGM_DATA/bolus_wizard_estimate.csv',sep=',',parse_dates=[2],index_col=2,na_values='nil')
day = df.ix['2010-10-04']
highs = day[day['glucose'] > 180]
lows = day[day['glucose'] < 80]

bolus = bolus_df.ix['2010-10-04']
bolus['bg_input'] = bolus['bg_input']/2.0
figure(figsize=(9,6));
ax = gca()
day['glucose'].plot(style='k--',ax=ax)
highs['glucose'].plot(style='ro',ax=ax)
lows['glucose'].plot(style='bo',ax=ax)

bolus['carb_input'].plot(style='go',ax=ax)
bolus['bg_input'].plot(style='mo',ax=ax,label='bolus')
legend(loc=2)
```
