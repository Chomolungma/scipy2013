### 2D Gaussian distribution centered at 1,1 with covariance of ([[1,0],[3,10]]

```python
points = np.random.multivariate_normal(
    mean=(1,1), cov=[[1, 0],[3, 10]], size=1000)

x, y = points.T
plt.plot(x, y, 'r.')
```

#Calculate Eigenvalues and EigenVectors (PCA)

```python
mean = np.mean(points,axis=0)
demeaned = points-mean
evals, evecs = np.linalg.eig(np.cov(demeaned.T))

#evals,evecs are not guaranteed to be ordered

order = evals.argsort()[::-1]
print evals[order]
print evecs[:,order[0]]
print evecs
print evals/np.linalg.norm(evals)
```

### SVD Interface

```python
# Center data
X = points
mean_ = np.mean(X, axis=0)
X -= mean_
U, S, V = np.linalg.svd(X, full_matrices=False)
print V
print S
print S**2/points.shape[0]
print U
```

### SKLEAN INTERFACE

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, whiten=False).fit(points)
print pca.components_
print pca.explained_variance_ratio_
print pca.explained_variance_
```
