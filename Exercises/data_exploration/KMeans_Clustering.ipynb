#Returning back to PCA
##Goal: Reduce dimensionality!

```python
from sklearn.decomposition import PCA
```

```python
df.columns
```

```python
features = df[df.columns[:4]]
```

```python
pca = PCA(n_components=2, whiten=True).fit(features)
```

```python
#eigenvectors
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
#amount of coverage from pca
pca.explained_variance_ratio_.sum()  
```

```python
#project data in pca space (equivalent of np.dot(eigenvector,data)
X_pca = pca.transform(features)
```

#Use KMeans on PCA

* KMeans: partitions features into K clusters
*  Can be difficult to determine number of clusters

```python
from sklearn.cluster import KMeans
from numpy.random import RandomState
rng = RandomState(73)
```

```python
kmeans = KMeans(3, random_state=rng).fit(X_pca)
```

```python
kmeans.cluster_centers_
```

#Visualize the clusters

```python
import pylab as pl
from itertools import cycle
def plot_2D(data, target, target_names,title):
    colors = cycle('rgbcmykw')
    target_ids = range(len(target_names))
    pl.figure()
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 0], data[target == i, 1],
                   c=c, label=label)
    pl.title(title)
    pl.legend()
    pl.show()
```

```python
plot_2D(X_pca,kmeans.labels_, list(set(df.types.values)),title='Kmeans with PCA')
```
